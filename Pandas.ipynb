```python
# This is a sample pandas scfript how to group columns in a dataframe and get the aggreate counts

import pandas as pd
import matplotlib.pyplot as plt

fn="/content/sample_data/Electric_Vehicle_Population_Data.csv"
df = pd.read_csv(fn)
state_counts = df.groupby('Make').size()
print(state_counts)
state_counts = df['Make'].value_counts()

plt.xticks(rotation=60,fontsize=6)
plt.bar(state_counts.index, state_counts.values)
plt.xlabel("Make")
plt.ylabel("Number of Vehicles")
plt.title("Number of Vehicles by Company")
# Set the figure size
fig = plt.gcf()
fig.set_figheight(5)
fig.set_figwidth(12)
plt.show()
```
